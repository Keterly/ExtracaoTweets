# Autentica√ß√£o

```python
with open('keys.txt', 'r') as file:
    access_token = file.readline().strip('\n')
    access_token_secret = file.readline().strip('\n')
    consumer_key = file.readline().strip('\n')
    consumer_secret = file.readline().strip('\n')
```

# Listando tweets da timeline

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

public_tweets = api.home_timeline()
for tweet in public_tweets:
    print(tweet.text)
```

# Extraindo tweets sobre #bbb22

```python
tweets = tweepy.Cursor(api.search_tweets, q='#bbb22').items(1000)
```

```python
tweets_data = []
```

```python
for tt in tweets:
    tts = re.sub("https+", " ", tt.text).lower()
    tts = re.sub("rt", " ", tts)
    tts = re.sub('://t.co/+', " ", tts)
    tweets_data.append(tts)
    print(tts)
```

```python
data = {'tweets': tweets_data}
```

```python
data
```

```python
df = pd.DataFrame(data)
```

# Nuvem de palavras - O que est√£o falando sobre o BBB 22?

```python
text = " ".join(tweet.lower() for tweet in df['tweets'])
```
